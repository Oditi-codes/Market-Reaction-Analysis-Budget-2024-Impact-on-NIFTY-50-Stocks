# **Market Reaction Analysis: Budget 2024 Impact on NIFTY 50 Stocks**

# **Problem Statement**
**Analysing how India's 2024 Union Budget announcement impacted stock market behavior, using NIFTY 50 companies as a sample.**

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'ITC.NS', 'HDFCBANK.NS', 'MARUTI.NS']
data = yf.download(stocks, start='2024-01-20', end='2024-02-10')['Close']
```

```python
print(type(data))
print(data.head())
```

#Statistics

ðŸ”¹ 1. Daily Return

Definition:
Daily return represents the percentage change in a stockâ€™s price from one trading day to the next. It helps us understand the short-term movement â€” whether the stock went up or down on a particular day.

Formula:
Daily Returnt={Price(t)âˆ’Price(tâˆ’1)}/Price(tâˆ’1)

Example:
If TCS was â‚¹3720 on Jan 24 and â‚¹3750 on Jan 25,
Daily Return=(3750âˆ’3720)/3720=0.00806=0.81%


ðŸ”¹ 2. Mean Return

Definition:
Mean return is the average of daily returns over a given period. It helps us see if a stock is generally trending upward (positive mean) or downward (negative mean).

Why it's useful:
It gives a quick snapshot of profitability or consistent growth/decline over the time window being analyzed.

ðŸ”¹ 3. Volatility

Definition:
Volatility is the standard deviation of daily returns. It measures how much the stock price fluctuates â€” i.e., the risk or instability in its behavior.

Interpretation:

    High volatility: More price swings â†’ higher risk, potentially higher reward.

    Low volatility: More stable performance â†’ safer but may yield lower returns.

```python
daily_returns = data.pct_change().dropna()
mean_returns = daily_returns.mean()
volatility = daily_returns.std()
```

```python
# Price Trends
data.plot(figsize=(12,6), title="Stock Prices Before and After Budget 2024")
plt.ylabel("Adjusted Closing Price")
plt.grid(True)
```

```python
# Bar charts
mean_returns.sort_values().plot(kind='barh', title='Average Daily Return')
plt.grid(True)

volatility.sort_values().plot(kind='barh', title='Volatility (Std Dev of Returns)')
plt.grid(True)
```

```python
sns.heatmap(daily_returns.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Stock Returns")
```

#Insights

*  TCS and INFY showed the highest positive correlation (0.88), suggesting sectoral coherence.
*  HDFCBANK had the lowest volatility (1.2%), making it the most stable during Budget 2024.
*  RELIANCE had the best average return post-budget at 1.8%.
